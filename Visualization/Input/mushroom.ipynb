```python
import matplotlib.pyplot as plt
```

### For all groups and percentage is 80%

```python
# x-coordinates of group id
left = [1, 2, 3, 4, 5, 6]
  
# heights of bars
height = [39,27,14,26,24,29] 
  
# labels for bars
tick_label = ['gone', 'two', 'three', 'four', 'five','soz']
  
# plotting a bar chart
plt.bar(left, height, tick_label = tick_label,
        width = 0.8, color = ['red', 'green'])
  
# naming the x-axis
plt.xlabel('x - axis - groups')
# naming the y-axis
plt.ylabel('y - axis - no. of attributes')
# plot title
plt.title('My bar chart!')
  
# function to show the plot
plt.show()
```
